```python
import os
import json

# Path to the feedback file
feedback_file = "user_feedback.json"

# Function to load existing feedback or initialize an empty file
def load_feedback():
    if os.path.exists(feedback_file):
        with open(feedback_file, 'r') as file:
            try:
                return json.load(file)
            except json.JSONDecodeError:
                return {}  # Start fresh if the file is invalid
    return {}  # Return empty if the file doesn't exist

# Function to save user feedback
def save_user_feedback(user_id, feedback_data):
    all_feedback = load_feedback()  # Load existing feedback
    # Ensure user-specific feedback exists
    if user_id not in all_feedback:
        all_feedback[user_id] = []
    # Append the new feedback
    all_feedback[user_id].append(feedback_data)
    # Save back to the JSON file
    with open(feedback_file, 'w') as file:
        json.dump(all_feedback, file, indent=4)

# Example interaction for collecting feedback
user_id = "user123"  # Example user ID
# Example inputs
weather = 5 # Degrees Celsius
activity = "casual"
mood = "energetic"
user_id = "user_123"  # Example user ID
structured_outfits = recommend_structured_outfits(weather, activity, mood)

# Collect feedback
for i, outfit in enumerate(structured_outfits, 1):
    print(f"Outfit {i}: {', '.join(outfit)}")
    while True:
        feedback = input(f"Did you like Outfit {i}? (y/n): ").strip().lower()
        if feedback in ['y', 'n']:
            break
        print("Invalid input. Please enter 'y' or 'n'.")

    # Prepare feedback data
    feedback_data = {
        "outfit_id": i,
        "items": outfit,
        "liked": feedback == 'y',
        "disliked": feedback == 'n'
    }
    save_user_feedback(user_id, feedback_data)

print("Feedback saved!")
```

```python
from segment_anything import sam_model_registry, SamPredictor

# Specify the model type and checkpoint path
model_type = "vit_h"  # Options: vit_b (Base), vit_l (Large), vit_h (Huge)
checkpoint_path = "sam_vit_h_4b8939.pth"  # Path to the downloaded SAM model checkpoint

# Load the SAM model
sam = sam_model_registry[model_type](checkpoint=checkpoint_path)
predictor = SamPredictor(sam)
```

```python
import cv2
import numpy as np

image_path = r"C:\Users\insha\cluless\Data\WOMEN\categories\Dresses\WOMEN-Dresses-id_00000023-01_1_front.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

predictor.set_image(image)
```

```python
from segment_anything import SamAutomaticMaskGenerator

mask_generator = SamAutomaticMaskGenerator(sam)
masks = mask_generator.generate(image)
```

```python
import matplotlib.pyplot as plt

image_with_masks = image.copy()
for mask in masks:
    mask_overlay = mask["segmentation"].astype(np.uint8)*255
    contours,_ = cv2.findContours(mask_overlay, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(image_with_masks, contours, -1, (255, 0, 0), 2)

plt.figure(figsize=(10, 10))
plt.imshow(image_with_masks)
plt.axis('off')
plt.title("Image with Segmentation Masks")
plt.show()
```

```python
# Get the dimensions
height, width, _ = image.shape  # height, width, channels

# Calculate resolution
resolution = width * height
print(f"Width: {width}, Height: {height}, Resolution: {resolution}")
```

```python
# Set a minimum size threshold for masks
min_mask_area = 50000  # Adjust based on your image resolution
#we're starting with 50,000 cause our img resolution is 825750
filtered_masks = []
for mask in masks:
    mask_area = np.sum(mask["segmentation"])
    if mask_area > min_mask_area:
        filtered_masks.append(mask)
```

```python
image_height, image_width = image.shape[:2]

clothing_masks = []
for mask in filtered_masks:
    y_coords, x_coords = np.where(mask["segmentation"])
    if y_coords.mean() > image_height * 0.3:  # Ignore masks near the top (face/head)
        clothing_masks.append(mask)
```

```python
final_clothing_masks = []
for mask in clothing_masks:
    y, x, w, h = cv2.boundingRect(mask["segmentation"].astype(np.uint8))
    aspect_ratio = h / w
    if 0.5 < aspect_ratio < 3.0:  # Clothing tends to have a reasonable aspect ratio
        final_clothing_masks.append(mask)
```

```python
selected_mask = final_clothing_masks[0]["segmentation"]

# Create a binary mask
binary_mask = selected_mask.astype(np.uint8) * 255

# Crop the clothing using the refined mask
x, y, w, h = cv2.boundingRect(binary_mask)
clothing_object = image[y:y+h, x:x+w]

# Display the cropped clothing object
plt.figure(figsize=(5, 5))
plt.imshow(clothing_object)
plt.axis('off')
plt.title("Cropped Clothing Object")
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np

def find_dominant_color(image, mask, k=3):
    # Apply mask to the image
    clothing_region = cv2.bitwise_and(image, image, mask=mask)
    
    # Reshape the region into a 2D array of pixels
    pixels = clothing_region.reshape((-1, 3))  # (N, 3)
    pixels = pixels[np.sum(pixels, axis=1) > 0]  # Remove black pixels (background)
    
    if len(pixels) == 0:
        raise ValueError("No valid pixels found in the clothing region.")
    
    # Apply K-Means clustering
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(pixels)
    
    # Get the most frequent cluster
    labels, counts = np.unique(kmeans.labels_, return_counts=True)
    dominant_label = labels[np.argmax(counts)]
    
    # Get the dominant color (cluster center)
    dominant_color = kmeans.cluster_centers_[dominant_label]
    return dominant_color.astype(int)

# Example usage
try:
    dominant_color = find_dominant_color(image, binary_mask)
    print("Dominant Color:", dominant_color)  # Output: [R, G, B]
except ValueError as e:
    print(e)
```

```python
import webcolors

def get_color_name(rgb_color):
    try:
        # Try to find the exact color name
        color_name = webcolors.rgb_to_name(rgb_color)
    except ValueError:
        # If no exact match, find the closest color name
        closest_name = None
        min_distance = float('inf')
        for name, standard_rgb in webcolors.CSS3_NAMES_TO_HEX.items():
            r_std, g_std, b_std = webcolors.hex_to_rgb(standard_rgb)
            distance = ((r_std - rgb_color[0]) ** 2 +
                        (g_std - rgb_color[1]) ** 2 +
                        (b_std - rgb_color[2]) ** 2)
            if distance < min_distance:
                min_distance = distance
                closest_name = name
        color_name = closest_name
    return color_name

# Example usage
rgb_color = dominant_color  # [R, G, B] from the function
color_name = get_color_name(tuple(rgb_color))
print("Dominant Color Name:", color_name)
```
